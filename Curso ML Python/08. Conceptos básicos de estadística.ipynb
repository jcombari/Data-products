```python
import pandas as pd
import numpy as np
from scipy.stats import mode
from IPython.display import Image
import matplotlib.pyplot as plt

%matplotlib inline
```

## Medidas de centralización

### Media

La media es el promedio de un conjunto de números

```python
np.mean(X)
```

```python
np.sum(X) / len(X)   # Cálculo aplicando la fórmula de la media  aritmética
```

### Mediana

Representa el valor de la variable de posición central en un conjunto de datos ordenados.

```python
np.median(X)
```

### Moda

La moda es el valor con mayor frecuencia en una distribución de datos.

```python
mode(X)[0][0]
```

### Percentil

Una vez ordenados los datos de menor a mayor, el percentil indica el valor de la variable por debajo del cual se encuentra un porcentaje dado de observaciones en un grupo de observaciones. Por ejemplo, el percentil 20º es el valor bajo el cual se encuentran el 20 por ciento de las observaciones

```python
np.percentile(X, 50)   # Percentil 50, equivalente a la media
```

```python
np.percentile(X, 25)  # Percentil 25
```

## Medidas de dispersión

Nos indican cómo de dispersos están los datos. La varianza y la desviación típica nos indican cómo se dispersan los datos respecto de la media.

### Varianza

```python
np.var(X)
```

### Desviación típica

```python
np.std(X)
```

### Coeficiente de variación

Mide la variabilidad relativa entre la media y la desviación típica

```python
np.std(X) / np.mean(X)
```

## Medidas de asimetría

```python
plt.hist(X)
```

Podemos comprobar que es ligeramente con asimetría positiva y tiende a ser leptocúrtica
